```python
# Import expediture features
expenses = pd.read_csv('https://download.bls.gov/pub/time.series/cx/cx.series',delimiter = "\t")
```

```python
expenses.head()
```

```python
cat_codes = ['EXPEND','INCOME']
expenses_c = expenses.loc[(expenses['demographics_code'] == 'LB11') &
                          (expenses['characteristics_code'] == '01')]

expenses_c = expenses_c[expenses_c.category_code.isin(cat_codes)]

expenses_c.tail()
```

```python
series = expenses_c['series_id'].tolist()
expenses_key = expenses_c[['series_id','subcategory_code']]
expenses_key.head()
```

```python
# Import data file
df = pd.read_csv('https://download.bls.gov/pub/time.series/cx/cx.data.1.AllData', delimiter = "\t")

df.head()
```

```python
df_expense = df[df.series_id.isin(series)]

df_expense.head()
```

```python
df_expense_c = pd.merge(df_expense, expenses_key, on='series_id')
df_expense_c.head()
```

```python
df_expense_c = df_expense_c.drop(['footnote_codes','period'], axis=1)
df_expense_c.head()
```

```python
df_expense_c = df_expense_c[['year','subcategory_code','value']]

df_expense_c.tail()
```

```python
df_expense_c = df_expense_c.groupby(['year','subcategory_code']).sum().unstack('subcategory_code')

df_expense_c.head()
```
